```python
def fill_with_prev_avg(row):
    for i in range(len(row)):
        if pd.isna(row[i]) and i >= 3:
            row[i] = row[i-3:i].mean()
    return row

# Apply row-wise
d_dec25 = d_dec25.apply(fill_with_prev_avg, axis=1)



# Create new column with row-wise average (excluding first column)
d_dec25["Average"] = d_dec25.iloc[:, 1:].mean(axis=1)
#month number
d_dec25["Month"]=12
#one hot encoding
d_dec25["Delhi"]=1
d_dec25["Gwalior"]=0
d_dec25["Jabalpur"]=0
#dropping columns
d_dec_preprocess = d_dec25.drop(columns=d_dec25.loc[:, "00:00:00":"23:00:00"].columns)


print(d_dec_preprocess)
```
